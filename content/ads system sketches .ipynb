```python
import numpy as np
import pandas as pd

# Simulation parameters
np.random.seed(42)
num_days = 7
impressions_per_day = 1000
ads_per_auction = 5

# Assume each ad's conversion value is always 1
conversion_value = 1.0

# For simplicity, assume we have a pool of ads with fixed bidding strategies
num_ads = 20
# Each ad has a "bid" that equals its probability of conversion
# Since model is calibrated, probability of conversion == bid
ad_bids = np.random.uniform(0.0, 1.0, size=num_ads)

# We'll simulate day-over-day:
# For each day, we run multiple "auctions":
# 1. Select a subset of ads at random to participate in an auction.
# 2. The highest bid wins and is shown.
# 3. With probability = the winning bid, we get a conversion.
# We'll track total conversions, and the resulting ads score.

day_results = []
for day in range(num_days):
    print(f"Day {day+1}")
    # For each day, run the auctions
    total_conversions = 0
    for i in range(impressions_per_day):
        if i % 100 == 0:
            print(f"\tImpression {i+1} out of {impressions_per_day}")
        # Randomly pick a set of ads to participate
        candidate_indices = np.random.choice(num_ads, size=ads_per_auction, replace=False)
        candidate_bids = ad_bids[candidate_indices]

        # First-price auction: highest bid wins
        winner_idx = candidate_indices[np.argmax(candidate_bids)]
        winner_probability = ad_bids[winner_idx]
        sorted_bids = np.sort(candidate_bids)
        winner_bid = sorted_bids[-2]  # Second highest bid
        # Probability of conversion = bid
        
        conversion_happened = np.random.rand() < winner_probability
        if conversion_happened:
            # Ads score increments by conversion_value
            total_conversions += conversion_value

    # Store the day's total conversions (ads score for that day)
    day_results.append({"day": day+1, "ads_score": total_conversions})

results_df = pd.DataFrame(day_results)
display(results_df)

# Now 'results_df' has the day-wise ads score.
# This simplistic simulation shows how one might compute ads score day-over-day.
```
